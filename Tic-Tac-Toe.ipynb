```python
#function that creates the board
from IPython.display import clear_output
def display_board(board):
    clear_output()
    print(board[7]+'|'+board[8]+'|'+board[9])
    print('-----')
    print(board[4]+'|'+board[5]+'|'+board[6])
    print('-----')
    print(board[1]+'|'+board[2]+'|'+board[3])      
```

```python
#function that assigns the player's marker
def player_input():
    marker=''
    player2=''
    acceptable_choice='X','O'
    
    while marker not in acceptable_choice:
        marker=input('Player1: please chose your marker (X or O)=').upper()
        
        if marker not in acceptable_choice:
            print('Sorry, that is not a valid marker! Please chose again (X or O)')
    
    player1 = marker
    if player1 == 'X':
        player2 = 'O'
    else:
        player2 = 'X'
     
    return player1,player2
```

```python
#function that randomly decides which player goes first
import random
def choose_first():
    return random.randint(1,2)

def space_check(board,position):
    position=int(position)
    return board[position]==' '
```

```python
#function that places the player's marker on the board
def place_marker(board, marker, position):   
    board[position]=marker
    return display_board(board)
```

```python
#function that checks if the board is full and returns a boolean
def full_board_check(board):
    full=True
    for board_positions in board:
        if (board_positions.count(' ')>=1):
            full=False
            break
    return full
```

```python
#function that asks for the player's next move (as a number 1-9) and then checks if the position is free
def player_choice(board):
    move='modify'
    acceptable_range=range(0,10)
    whitin_range=False
    
    while move.isdigit() == False or whitin_range == False:
        move=input('Next player please chose a position as a number from 1 to 9:')
        
        if move.isdigit() == False:
            print('Sorry, that is not a digit. Please choose from 1 to 9.')
        
        if move.isdigit() == True:
            if int(move) in acceptable_range and space_check(board,int(move)) == True:
                whitin_range = True   
            else:
                 print('Sorry, that position is not valid. Please choose another one.')
        
    return int(move)
```

```python
#function that checks if the player has won
def win_check(board, mark):
    if board[7]==board[5]==board[3]==mark or board[1]==board[5]==board[9]==mark or board[1]==board[4]==board[7]==mark:
        return True
    elif board[2]==board[5]==board[8]==mark or board[3]==board[6]==board[9]==mark or board[1]==board[2]==board[3]==mark:
        return True
    elif board[4]==board[5]==board[6]==mark or board[7]==board[8]==board[9]==mark:
        return True
    else:
        return False
```

```python
#function that asks if the player wants to play again and returns a boolean
def replay():
    acceptable_choice=['Y','N']
    play_again='asdf'
    
    while play_again not in acceptable_choice:
        play_again=input('Do you want to play again? Enter Y(Yes) or N(No)):').upper()
        
        if play_again not in acceptable_choice:
            print('That is not a valid answer. Please pick Y or N')
    
    return play_again
```

```python
#GAME SCRIPT
print('Welcome to Tic Tac Toe!')
game_on = 'Y'
board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
player1_marker=''
player2_marker=''
is_over=False
winner1=False
winner2=False

while game_on == 'Y':
    player1_marker, player2_marker = player_input()
    print(f"This is Player's 1 marker:{player1_marker} and this is Player's 2 marker:{player2_marker}")    
    
    who_starts=choose_first()
    print(f'Player {who_starts} goes first')
    
    

    #primeira jogada
    position=int(input(f'Player {who_starts} please choose a position as a number from 1 to 9:'))
    display_board(board)
    if who_starts==1:
        place_marker(board,player1_marker,position)
    else:
        place_marker(board,player2_marker,position)
    
    #segunda jogada
    next_move=player_choice(board)
    if player1_marker in board:
        place_marker(board,player2_marker,next_move)
    else:
        place_marker(board,player1_marker,next_move)
    
    
    #proximas jogadas
    while is_over == False and winner1 == False and winner2 == False:
        
        following_move=player_choice(board)
        
        if board.count(player1_marker) == board.count(player2_marker):
            if who_starts == 1:  
                place_marker(board,player1_marker,following_move)
            else:
                place_marker(board,player2_marker,following_move)
                
        else:
            if who_starts == 1:
                place_marker(board,player2_marker,following_move)
            else:
                place_marker(board,player1_marker,following_move)
        
        is_over = full_board_check(board)
        
        winner1 = win_check(board, player1_marker)
        winner2 = win_check(board, player2_marker)
        
    
    #quem ganhou?
    if win_check(board, player1_marker) == True:
        print (f'{player1_marker} was won!')
    
    elif win_check(board, player2_marker) == True:
        print (f'{player2_marker} was won!')
    else:
        print ('Its a tie! Nobody has won the game')
    
    #jogar de novo?
    game_on = replay()
    if player1_marker in board:
        place_marker(board,player2_marker,next_move)
    else:
        place_marker(board,player1_marker,next_move)
    
    
    #proximas jogadas
    while is_over == False and winner1 == False and winner2 == False:
        
        following_move=player_choice(board)
        
        if board.count(player1_marker) == board.count(player2_marker):
            if who_starts == 1:  
                place_marker(board,player1_marker,following_move)
            else:
                place_marker(board,player2_marker,following_move)
                
        else:
            if who_starts == 1:
                place_marker(board,player2_marker,following_move)
            else:
                place_marker(board,player1_marker,following_move)
        
        is_over = full_board_check(board)
        
        winner1 = win_check(board, player1_marker)
        winner2 = win_check(board, player2_marker)
        
    
    #quem ganhou?
    if win_check(board, player1_marker) == True:
        print (f'{player1_marker} was won!')
    
    elif win_check(board, player2_marker) == True:
        print (f'{player2_marker} was won!')
    else:
        print ('Its a tie! Nobody has won the game')
    
    #jogar de novo?
    game_on = replay()
```
